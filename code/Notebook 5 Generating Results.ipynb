```python
import pandas as pd
import matplotlib.pyplot as plt


busydf=pd.read_csv('../preprocessed data/busiest_locations.csv', sep=',',index_col = 0)

RevPerZonedf=pd.read_csv('../preprocessed data/locationTotalamountMean.csv', sep=',',index_col = 0)

lookupdf = pd.read_csv('../raw_data/taxi_zone_lookup.csv')
```

# Merging Data

```python
#Merging the 2 revenue per zone and number of pickups dfs

Result_data = busydf.merge(RevPerZonedf, left_on='Pickup Location ID', right_on='PULocationID')

#removing duplicate column
Result_data = Result_data.drop(labels="PULocationID",axis = 1)

Result_data
```

```python
#Merging lookup file with the result table
Named_data = lookupdf.merge(Result_data, left_on='LocationID', right_on='Pickup Location ID')

Named_data = Named_data.drop(labels=['Pickup Location ID','service_zone'],axis = 1)
Named_data
```

# Sorting values

```python
#Observing the top 20 busiest locations
Named_data.sort_values(by = 'Number of Pickups',ascending = False).head(20)
```

```python
#Observing the locations with the  top 20 highest average total_amounts
Named_data.sort_values(by = 'total_amount',ascending = False).head(20)
```

```python
#Sorting the top 20 busiest locations by most expensive 

NUMBER_OF_MONTHS = 6

NOpickupsDF = Named_data.sort_values(by = 'Number of Pickups',ascending = False).head(20)

#Extracting the top 10 locations
NOpickupsDF = NOpickupsDF.sort_values(by = ['total_amount'],ascending = False).head(10)


#Finding the number of pickups per month rounding down to approximate
NOpickupsDF['Number of Pickups']=NOpickupsDF.apply(lambda x: int(x['Number of Pickups']/NUMBER_OF_MONTHS),axis=1)


NOpickupsDF


```

```python
#Sorting the top 20 most expensive zones by number of pickups
totalAVcostDF = Named_data.sort_values(by = 'total_amount',ascending = False).head(20)


#Extracting the top 10 locations
totalAVcostDF = totalAVcostDF.sort_values(by = ['Number of Pickups'],ascending = False).head(10)

#Rounding down to approximate number
totalAVcostDF['Number of Pickups']=totalAVcostDF.apply(lambda x: int(x['Number of Pickups']/NUMBER_OF_MONTHS),axis=1)
#Highest trips zones are all in and around Queen's which is Near JFK airport

totalAVcostDF
#note Queens is most fequent in both and JFK is top in both
```
